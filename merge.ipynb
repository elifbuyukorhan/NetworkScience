```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
bio_graph = nx.read_edgelist('C:\\Users\\ElifB\\Desktop\\bio-yeast.txt', comments='#', nodetype = int)
num_nodes = len(nx.nodes(bio_graph))
num_edges = len(nx.edges(bio_graph))
print("Number of nodes: ", num_nodes, "\nNumber of edges: ", num_edges)
```

```python
G_as = nx.Graph()
G_disas = nx.Graph()
G_n = nx.erdos_renyi_graph(num_nodes, 0.001834)
```

```python
# Get the degree of each node
degrees = dict(bio_graph.degree())
```

```python
while(len(list(bio_graph.edges()))!=0):
    p = 0.5
    edge1 = random.choice(list(bio_graph.edges()))
    n1, n2 = edge1
    edge2 = random.choice(list(bio_graph.edges()))
    n3, n4 = edge2
    nodes = (n1,n2,n3,n4)
    k = (degrees[n1], degrees[n2], degrees[n3], degrees[n4])
    k_sorted = sorted(k)
    first_n = 0
    second_n = 0
    third_n = 0
    fourth_n = 0
    for n in nodes:
        if(k_sorted[0]==degrees[n] and second_n!=n and third_n!=n and fourth_n!=n and first_n == 0):
            first_n = n
        if(k_sorted[1]==degrees[n] and first_n!=n and third_n!=n and fourth_n!=n and second_n == 0):
            second_n = n
        if(k_sorted[2]==degrees[n] and first_n!=n and second_n!=n and fourth_n!=n and third_n == 0):
            third_n = n
        if(k_sorted[3]==degrees[n] and first_n!=n and second_n!=n and third_n!=n and fourth_n == 0):
            fourth_n = n

    bio_graph.remove_edges_from([(n1,n2),(n3,n4)])
    x = random.random()
    control1 = G_as.has_edge(first_n,second_n)
    if not control1 and x <= p:
        G_as.add_edge(first_n,second_n)

    control2 = G_as.has_edge(third_n,fourth_n)
    if not control1 and x <= p:
        G_as.add_edge(third_n,fourth_n)
```

```python
# Compute the degree correlation matrix
D = nx.degree_mixing_matrix(G_as)
# Plot the degree correlation matrix
fig, ax = plt.subplots(figsize=(10, 10))
im = ax.imshow(D, cmap='Reds')
ax.set_xticks(range(len(D)))
ax.set_yticks(range(len(D)))
ax.set_xticklabels(range(len(D)))
ax.set_yticklabels(range(len(D)))
plt.xlabel('k1')
plt.ylabel('k2')
plt.title('The Degree Correlation Matrix Graph')
plt.colorbar(im)
plt.show()
```
